```python
import pandas as pd
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import math
import time
```

```python
x = 2*np.random.rand(2000,2)-1
y = sp.sqrt(x[:,0]**2 + x[:,1]**2)
x1 = x[y<=1]; x2 = x[y>1]
```

```python
plt.plot(x1[:,0],x1[:,1], 'ro')
plt.plot(x2[:,0],x2[:,1], 'o')
plt.axis([-1, 1, -1, 1])
plt.show()
```
